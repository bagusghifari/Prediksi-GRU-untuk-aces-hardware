```python
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl import Workbook
```

```python
# Membaca file CSV
df = pd.read_csv('imputed 2007-2023 (1).csv')
```

```python
# Mengubah kolom 'Date' menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Mengatur kolom 'Date' sebagai index
df.set_index('Date', inplace=True)
```

```python
# Menghitung moving average dari kolom 'close'
df['MA'] = df['close'].rolling(window=15).mean()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['close'], label='Close Price')
plt.plot(df.index, df['MA'], label='15-Day Moving Average', color='orange')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price and 15-Day Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Menyimpan hasil ke file CSV baru
df.to_csv('aces_ma15.csv', index=True)

print(df.head(30))
```
